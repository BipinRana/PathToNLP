#### Entropy quantifies the randomness or unpredictability in a system, helping to measure how mixed or pure the data is. 
#### Higher entropy means more disorder, while lower entropy indicates more certainty or uniformity.

```python
import numpy as np
```

```python
x = np.linspace(0.001,0.999,100)
```

```python
y = -x * np.log2(x) - (1 - x) * np.log2(1 - x)  # Entropy formula
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.title('Entropy in Boolean Data')
plt.xlabel("Probability of Yes")
plt.ylabel("Entropy")
```

#### From this figure, we can deduce that entropy is highest when the probability of "Yes" is around 0.5, 
#### meaning the data is most uncertain when both outcomes are equally likely. 
#### As the probability moves closer to 0 or 1, entropy decreases, indicating lower uncertainty and a more predictable outcome.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
# Generate random scatter points

plt.scatter(X, y, color='blue', label="Random Points")  # Scatter plot

# Vertical line at x = -1
plt.axvline(x=-1, color='red', label="First Decision Node")  

# Horizontal line at y = 2, but starting after x = -1
plt.plot(np.linspace(-1, 5, 100), [2] * 100, color='green', label="Second Decision Node")

# Labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Conditional Lines")
plt.legend()

plt.show()
```

#### *From this figure, we can interpret how a decision tree partitions the feature space using axis-aligned boundaries. 
#### *The red vertical line at =1 represents the first decision node, separating the data into two regions: 
#### *The left side belongs to Class 1, while the right side represents not Class 1. 
#### *The green horizontal line at =2 further divides the right region into two more classes: 
#### *The upper section corresponds to Class 2, while the lower section represents Class 3. 
#### *This illustrates how decision trees recursively split the data space into distinct decision regions.
