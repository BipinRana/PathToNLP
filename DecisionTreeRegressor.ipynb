```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
X = data.data
y = data.target
```

```python
import pandas as pd
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(f"Decision Tree\nMean Absolute Error: {mean_absolute_error(y_pred,y_test)}")
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=data.feature_names, filled=True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train,y_train)
```

```python
y_pred_linear = linear_model.predict(X_test)
```

```python
print(f"Linear Regression\nMean Absolute Error: {mean_absolute_error(y_pred_linear,y_test)}")
```
