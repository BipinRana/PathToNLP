```python
import pandas as pd
df = pd.read_csv('Bank-Customer-Attrition-Insights-Data.csv')
```

```python
df.isna()
```

```python
X = df[['CreditScore','Geography','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary','Satisfaction Score','Card Type','Point Earned']]
```

```python
X
```

```python
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first',sparse_output=False)
encoder.fit(X_train[['Geography','Gender']])
```

```python
X_train
```

```python
X_train[['is_from_Germany','is_from_Spain','is_Male']] = encoder.transform(X_train[['Geography','Gender']])
```

```python
X_train
```

```python
X_train.drop(['Geography','Gender'],axis=1,inplace=True)
```

```python
X_train
```

```python
X_test[['is_from_Germany','is_from_Spain','is_Male']] = encoder.transform(X_test[['Geography','Gender']])
```

```python
X_test.drop(['Geography','Gender'],axis=1,inplace=True)
```

```python
X_train['Card Type'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
Oencoder = OrdinalEncoder(categories=[['SILVER', 'GOLD', 'PLATINUM','DIAMOND']])
Oencoder.fit(X_train[['Card Type']])
X_train['Card Type'] = Oencoder.transform(X_train[['Card Type']])
X_test['Card Type'] = Oencoder.transform(X_test[['Card Type']])
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

LR = LogisticRegression()
LR.fit(scaled_X_train,y_train)
y_pred = LR.predict(scaled_X_test)

print('Logisitic Regression:')
print(classification_report(y_test,y_pred))


LR2 = LogisticRegression(penalty='l1', solver='liblinear', C=0.03)
LR2.fit(scaled_X_train,y_train)
y_pred2 = LR2.predict(scaled_X_test)

print('Logisitic Regression with Regularization (C = 0.03):')
print(classification_report(y_test,y_pred2))
pd.DataFrame(LR2.coef_,columns=X_train.columns)
```

```python
LR2 = LogisticRegression(penalty='l1', solver='liblinear', C=0.03,class_weight='balanced')
LR2.fit(scaled_X_train,y_train)
y_pred2 = LR2.predict(scaled_X_test)

print('Logisitic Regression with Regularization (C = 0.03) and adjusted class weights:')
print(classification_report(y_test,y_pred2))
pd.DataFrame(LR2.coef_,columns=X_train.columns)
```
