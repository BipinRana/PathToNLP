```python
import numpy as np
```

```python
x = np.linspace(0.001,0.999,100)
```

```python
entropy = -x * np.log2(x) - (1 - x) * np.log2(1 - x)  # Entropy formula
gini = 1 - (x**2+(1-x)**2)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,entropy,color='red',label='Entropy')
plt.plot(x,gini,color='green',label='Gini Impurity')
plt.title('Entropy v/s Gini Impurity')
plt.xlabel("Probability of Yes")
plt.ylabel("Entropy")
plt.legend()
plt.show()
```

### Gini vs. Entropy in Decision Trees

##### Both Gini Impurity and Entropy are criteria used to measure the purity of a node in decision trees:

#### Gini Impurity: Measures the probability of misclassifying a randomly chosen element. It is computationally faster since it avoids logarithms.
#### Entropy: Based on information theory, it measures the amount of information needed to classify an element. It can lead to slightly better splits in some cases but is computationally more expensive.

### When to use?

##### Use Gini when you want faster computations (default in CART).
##### Use Entropy if you prefer an information-theoretic approach (used in ID3 & C4.5).
##### In practice, both usually give similar results! ðŸš€

### Information Gain and Its Role in Branching
##### Information Gain (IG) is used in decision trees to decide which feature should be chosen for splitting at each node. 
##### It quantifies how much a feature improves the purity of the resulting subsets after a split.


### How It Works in Branching:

##### Split the Data on a Feature:Divide the data into subsets based on different values of a chosen feature.
