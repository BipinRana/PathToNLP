```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df_train[['bedrooms','grade','has_basement','living_in_m2','renovated','nice_view','perfect_condition','real_bathrooms','has_lavatory','single_floor','month','quartile_zone']])
scaled_X_train =  scaler.transform(df_train[['bedrooms','grade','has_basement','living_in_m2','renovated','nice_view','perfect_condition','real_bathrooms','has_lavatory','single_floor','month','quartile_zone']])
scaled_X_test = scaler.transform(df_test[['bedrooms','grade','has_basement','living_in_m2','renovated','nice_view','perfect_condition','real_bathrooms','has_lavatory','single_floor','month','quartile_zone']])
```

```python
#linear Regression 
import numpy as np

# Linear Regression model using gradient descent
def linear_regression(X, y, learning_rate=0.01, iterations=1000):
    m, n = X.shape
    W = np.zeros(n)  # Initialize weights
    b = 0  # Initialize bias
    for i in range(iterations):
        predictions = np.dot(X, W) + b
        errors = predictions - y
        # Compute gradients
        dw = (1 / m) * np.dot(X.T, errors)
        db = (1 / m) * np.sum(errors)
        # Update weights
        W -= learning_rate * dw
        b -= learning_rate * db
    return W, b

# Train the model
W, b = linear_regression(scaled_X_train, df_train['price'])
```

```python
prediction =np.dot(scaled_X_test, W)+b
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(df_test['price'],prediction)
print(f'The error is {error}')
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load dataset
iris = load_iris()
X = iris.data
y = (iris.target == 0).astype(int)  # Binary classification: Setosa vs Not Setosa

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Logistic Regression model using gradient descent
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def logistic_regression(X, y, learning_rate=0.01, iterations=1000):
    m, n = X.shape
    W = np.zeros(n)  # Initialize weights
    b = 0  # Initialize bias
    for i in range(iterations):
        z = np.dot(X, W) + b
        predictions = sigmoid(z)
        
        # Compute gradients
        dw = (1 / m) * np.dot(X.T, (predictions - y))
        db = (1 / m) * np.sum(predictions - y)
        # Update weights
        W -= learning_rate * dw
        b -= learning_rate * db
    return W, b

def threshold(x):
    return 1 if x>0.5 else 0
```

```python
# Train the model
W, b = logistic_regression(X_train, y_train)
z = np.dot(X_test, W) + b
predictions = sigmoid(z)
predict = np.vectorize(threshold)
predictions = predict(predictions)
```

```python
from sklearn.metrics import accuracy_score
error = accuracy_score(y_test,predictions)
print(f'The accuracy is {error}')
```
